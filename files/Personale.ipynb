```python
def read_date(stringa):
    elems = np.flip([ int(elem) for elem in stringa.split("/") ])
    return elems
```

```python
from datetime import date

class Author():
    
    def __init__(self,row):
        
        self.iris = row['\ufeffID persona (IRIS)']
        self.nome = row["Nome"].upper()
        self.cognome = row["Cognome"].upper()
        self.inizio = row["data di inizio affiliazione"]
        self.fine = row["data di fine affiliazione"]
        self.ruolo = row["ruolo"]
        self.qualifica = row["qualifica"]
        self.ssd = row["settore SSD"]
        self.orcid = row["ID persona (ORCID)"]
        self.scopus = row["ID persona (scopus)"]  
        self.pubs = []
        
        ssd = self.ssd
        self.ssd = ssd.split("-")[0][8:].strip()
        
        if self.fine == "":
            self.fine = date(2023,2,1)
        else:
            elems = read_date(self.fine)
            self.fine = date(*elems)

        elems = read_date(self.inizio)
        self.inizio = date(*elems)

        
        self.years = np.arange(self.inizio.year,self.fine.year+1)

        if self.orcid == "": self.orcid = None
        if self.scopus == "": 
            self.scopus = None
        else:
            self.scopus = self.scopus.split("=")[-1]
        
```

```python
import glob
import csv
files = glob.glob("list_*.csv")
dict_authors = {}
for file in files:
    key = file.split("_")[1:3]
    key = "-".join(key)
    rows = []
    with open(file, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            rows.append(row)

    authors = []
    for row in rows:
        author = Author(row)
        if author.ruolo in ["Docenti","Ricercatori"]:
            authors.append(author)

    dict_authors[key] = authors
```

```python
iris = []
for key,authors in dict_authors.items():
    iris += [author.iris for author in authors]
iris = np.unique(iris)
```

```python
now = {}
for ir in iris:
    for authors in dict_authors.values():
        for author in authors:
            if author.iris == ir:
                if ir in now:
                    author0 = now[ir]
                    author0.years = np.unique(np.append(author0.years,author.years))
                else:
                    now[ir] = author        
```

```python
class Pub():
    
    def __init__(self,row):
        
        self.title = row['\ufeff"Title"']
        self.authors_ids = row["Scopus Author Ids"].split("|")
        self.authors_ids = [id.strip() for id in self.authors_ids]
        self.year = int(row["Year"])
        self.source_type = row["Source type"]
        self.pub_type = row["Publication type"]
        self.ruoli = []
        self.authors = []
        self.ssds = []
```

```python
pubs = []
for row in rows:
    pub = Pub(row)
    if pub.pub_type != "" and pub.pub_type is not None:
        pubs.append(pub)
```

```python
for pub in pubs:
    for author in authors:
        id = author.scopus
        if id in pub.authors_ids:
            pub.ruoli.append(author.ruolo)
            pub.authors.append(author.cognome+", "+author.nome)
            pub.ssds.append(author.ssd)
    pub.ssds = np.unique(pub.ssds)
```

```python
yauthors = {}
ypubs = {}

for year in range(2020,2023):
    yauthors[year] = [author for author in authors if year in author.years]
    
for year,now in yauthors.items():
    ypubs[year] = []
    for pub in pubs:
        if year == pub.year:
            logy = False
            for author in now:
                if author.scopus in pub.authors_ids: 
                    logy = True
                    break
            if logy: ypubs[year].append(pub)
    print(year,len(yauthors[year]),len(authors),len(ypubs[year]))
```

```python
for year in range(2020,2023):
    
    for author in yauthors[year]:
        author.pubs = []
        id = author.scopus
        aut_pubs = []
        for k,pub in enumerate(ypubs[year]):
            if id in pub.authors_ids:
                aut_pubs.append(k)
        author.pubs = aut_pubs

    c=0
    for pub in ypubs[year]:
        ids = pub.authors_ids
        for author in yauthors[year]:
            if author.scopus in ids:
                c += 1
                break
                
    print(year,c)

    dict_ssds = {}
    for ssd in ssds:
        dict_ssds[ssd] = []

    for author in yauthors[year]:
        dict_ssds[author.ssd] += author.pubs

    for ssd in ssds:
        dict_ssds[ssd] = np.unique(dict_ssds[ssd])

    inter2 = []
    for i,ssd in enumerate(ssds):
        ps = dict_ssds[ssd]
        for j,ssd1 in enumerate(ssds):
            ps1 = dict_ssds[ssd1]
            if j >= i: continue
            inter2 += list(np.intersect1d(ps,ps1))
            #print(i,j,inter2)
    c2 = len(inter2)

    inter3 = []
    for i,ssd in enumerate(ssds):
        ps = dict_ssds[ssd]
        for j,ssd1 in enumerate(ssds):
            ps1 = dict_ssds[ssd1]
            if j >= i: continue
            inter2 = np.intersect1d(ps,ps1)
            for k,ssd2 in enumerate(ssds):
                ps2 = dict_ssds[ssd2]            
                if k >= j: continue
                inter3 += list(np.intersect1d(inter2,ps2))
    c3 = len(inter3)
    
    print("{0:6d} {1:6.2f} {2:6.2f}".format(year,100*c2/c,100*c3/c))
```
