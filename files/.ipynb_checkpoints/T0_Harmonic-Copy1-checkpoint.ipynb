## Driven harmonic oscillators

```{div} full-width

The driven harmonic oscillator is an useful tool in seismology because it can be used for describing either a simple mechanical seismometer or  the deformation of a building when they are subjected to ground motion.
```

### Seismometer

```{div} full-width

Let us consider a seismometer composed by an probe mass attached to a spring and embedded in a viscous dashpot as depicted in figure (??). Let us denote with $h_0$ the height of the probe mass at rest and consider the absolute coordinate of the probe mass $u$ as the sum of the ground motion $u_g$, the height at rest $h_0$ and its relative motion with respect to it $z$

$$
u(t) = u_g(t) + h_0 + z(t)
$$ (H0:1)

The elastic, $F_e$, and anaelastic, $F_a$, forces acting on the probe mass will be given by 

$$
F_e=-k\,z \qquad F_a =-c\,\dot{z}
$$ (H0:2)

From the second Newton's law

$$
F_e+F_a = m\,\ddot{u}
$$ (H0:3)

declined in the case of a machanical seismometer as the one depicted in figure (??), we thus obtain the following differential equation

$$
-k\,z(t) - c\,\dot{z}(t) = m\,\big(\ddot{u}_g(t)+\ddot{z}(t)\big)
$$ (H0:4)

where $m$ is the probe mass, $k$ is the elastic constant of the spring and $c$ is the anaelastic friction. We note that the left hand side is the sum of the total forces acting on the probe mass while the right-hand side is the inertial force, i.e., the mass by its total acceleration. 

By dividing eq. {eq}`H0:4` by the mass $m$ and defining the natural angular frequency $\omega_0$ and the damping ratio $\xi$ as follows

$$
\omega_0 = \sqrt{\frac{k}{m}} \qquad \xi = \frac{c}{2\,\omega_0}
$$ (H0:5)

we obtain

$$
\ddot{z}(t) + 2\,\omega_0\,\xi\,\dot{z}(t) + \omega_0^2\,z(t) = -\ddot{u}_g(t)
$$ (H0:6)

that is the classical form of a driven harmonic oscillator, where the left-hand side describes a damped harmonic oscilator and the right-hand side is the driving term that, in the present case, corresponds to minus the ground acceleration.

In the following we will also make use of the natural period of the harmonic oscillator, which is simply given by

$$
T_0 =  \frac{2\,\pi}{\omega_0} = 2\,\pi\, \sqrt{\frac{m}{k}}
$$ (H0:7)

```


### Building

````{div} full-width

Let su consider a building and denote with $u$ the absolute coordinate of its top, which can be seen as the sum of the ground motion (i.e., at the bottom of the building) and the building height $h$

$$
u(t) = u_g(t) + h(t) 
$$ (H0:8)

Due to deformation of the building $h$ will vary with time and so we also write $h$ as the sum of its height at rest $h_0$ and its relative displacement $z$

$$
h(t) = h_0 + z(t)
$$ (H0:9)

The ratio $z/h_0$ can be seen as a measure of the internal strain and, so, the elastic and anaelastic forces will be proportional to $z$ and its time derivative $\dot{z}$, respectively

$$
F_e = -\frac{\mu\,z}{h_0}
\qquad
F_a = -\frac{\eta\,\dot{z}}{h_0}
$$ (H0:10)

with $\mu$ and $\eta$ being the elastic modulus and effective viscosity of the building. In thi respect, from the comparison with eq. {eq}`H0:2`, the elastic constant and anaelastic friction of the building are inversely proportional to the building height: $c\propto h_0^{-1}$ and $c\propto h_0^{-1}$.

Similar to the seismometer, from the second Newton's law, we thus can write 

$$
-k\,z(t) - c\,\dot{z}(t) = m\,\big(\ddot{u}_g(t)+\ddot{z}(t)\big)
$$ (H0:11)

and recast it in the form of the driven harmonic oscillator

$$
\ddot{z}(t) + 2\,\omega_0\,\xi\,\dot{z}(t) + \omega_0^2\,z(t) = -\ddot{u}_g(t)
$$ (H0:12)


```{admonition} The natural period of buildings
:class: note

As a rule of thumb, we note that the the natural period of building is roughly propotional to the the number of floors $n$ in the building. Indeed, considering $h_0=n\,\Delta h$ anmd $m=n\,\Delta m$ with $\Delta h$ and $\Delta m$ being the height and the mass of each floors, we obtain

$$
T_0 = 2\,\pi\,\sqrt{\frac{m}{k}} = n\,\delta T
$$ (H0:13)

with $\delta T$ being the natural period of each floor

$$
\delta T=2\,\pi\,\sqrt{\frac{\delta m}{\delta h}}
$$ (H0:14)

In contrast, it can checked that the damping ratio $\xi$ does not vary with the number of floors and it usually ranges from $0.05$ to $0.1$ (or, in the woording of the  engineers, from $5\%$ to $10%$).

```



```{admonition} Relative or total acceleration?
:class: note

As shown by eqs. {eq}`H0:6` and {eq}`H0:12`, both the relative motion of the seismometer and of the building are governed by the same differential equation, having the minus ground acceleration as the driven term. When we will discuss the seismometer, we will focus on the link betweem the ground and the relative motions as the former and the latter are what we want retrive and can measure, respectively. When we will discuss the building, instead, we will be more interested in quantifying the total acceleration to which it is subjected, and we will recover it as the sum of the ground and relative acceleration.

```
````

## Damped Harmonic oscillator

```{div} full-width

The damped harmonic oscillator is a classic problem in mechanics. It describes the movement of a mechanical oscillator (e.g. spring pendulum) under the influence of a restoring force and friction. Let us thus set to zero the ground motion acceletarion and solve the homogenous differential equation with specified initial condition on the position and velocity at time = $t=0$

$$
\ddot{z}_\mathrm{h}(t) + 2\,\omega_0\,\xi\,\dot{z}_\mathrm{h}(t) + \omega_0^2\,z_\mathrm{h}(t) = 0
$$
$$
z_\mathrm{h}(0) = u_0
\qquad
\dot{z}_\mathrm{h}(0) = v_0
$$ (H0:15)


The homogeneous solution can be found starting from this trial form

$$
z_\mathrm{h}(t) = Z_+\,e^{\gamma_+\,t} + Z_-\,e^{\gamma_-\,t}
$$ (H0:16)

and it is straigforwad to show that the two exponents $\gamma_\pm$ are the solution of a degree-2 polynomial 

$$
Z_\pm\,\big(\gamma_\pm^2+2\,\omega_0\,\xi\,\gamma_\pm + \omega_0\big)\,e^{\gamma_\pm\,t} = 0
$$ (H0:17)

and read

$$
\gamma_\pm = -\omega_0\,\xi \pm \omega_0\,\sqrt{\xi^2-1}
$$ (H0:18)

```

### Underdamped harmonic oscillator


```{div} full-width

Within the assumption that $\xi<1$, the square root in the right-hand side of eq. {eq}`H0:18` is imaginary and the two exponents can be rewritten as folllows

$$
\gamma_\pm = -\lambda \pm i\,\omega_d
$$ (H0:19)

with

$$
\lambda = \omega_0\,\xi \qquad
\omega_d = \omega_0\,\sqrt{1-\xi^2}
$$ (H0:20)

In view of the fact that the real part of the exponents coincides, we can recast eq. {eq}`H0:16` as follows

$$
\begin{align}
z_\mathrm{h}(t) &= \big(Z_+\,e^{i\,\omega_d\,t} + Z_-\,e^{-i\,\omega_d\,t}\big)\,e^{-\lambda\,t} \\
                &= \big(C\,\cos(\omega_d\,t) + S\,\sin(\omega_d\,t)\big)\,e^{-\lambda\,t}
\end{align}
$$ (H0:21)

where $C$ and $S$ are two constants of integration related to the former by

$$
C = Z_+ + Z_-   \qquad
S=i\,\big(Z_+ - Z_-\big)
$$ (H0:22)

In light of this, the assumption of  $\xi<1$ leads to underdamped oscillations: the system oscillates (with a slightly different frequency than the natural frequency) with the amplitude gradually decreasing to zero. The angular frequency of the underdamped harmonic oscillator is $\omega_d$  and the exponential decay of the underdamped harmonic oscillator is $\lambda$. 

The cases $\xi=1$ and $\xi>1$, instead, describe critical and overdamped oscillations. They are not relevant in seismology and we do not discuss them.

```

### Initial conditions

```{div} full-width

Considering the first order derivative of eq. {eq}`H0:21`

$$
\dot{z}_\mathrm{h}(t) = \big[-(C\,\omega_d+S\,\lambda)\,\sin(\omega_d\,t) + (S\,\omega_d-C\,\lambda)\,\cos(\omega_d\,t)\big]\,e^{-\lambda\,t} 
$$ (H0:23)

from eq. {eq}`H0:15` we have 

$$
z_\mathrm{h}(0)  = C  = u_0
\qquad \dot{z}_\mathrm{h}(0)  = S\,\omega_d - C\,\lambda  = v_0
$$ (H0:24)

This system of equations is solved by

$$
C = u_0 \qquad 
S = \frac{u_0\,\lambda + v_0}{\omega_d} 
$$ (H0:25)


and eq. {eq}`H0:21` and its derivatives become

$$
\begin{align}
& z_\mathrm{h}(t) = \left(u_0\,\cos(\omega_d\,t) +\frac{u_0\,\lambda  + v_0}{\omega_d}  \,\sin(\omega_d\,t)\right)\,e^{-\lambda\,t}
\\
& \dot{z}_\mathrm{h}(t) = \left(v_0  \,\cos(\omega_d\,t) - \frac{u_0\,\omega_0^2  + v_0\,\lambda}{\omega_d}\,\sin(\omega_d\,t)\right)\,e^{-\lambda\,t}
\\
&\ddot{z}_\mathrm{h}(t) = \left[a_0\,\,\cos(\omega_d\,t)-\frac{v_0\,\omega_0^2 + a_0\,\lambda }{\omega_d}  \,\sin(\omega_d\,t)\right]\,e^{-\lambda\,t}
\end{align}
$$ (H0:26)

with $a_0$ being the initial acceleration

$$
a_0 = -\big(u_0\,\omega_0^2  + 2\,v_0\,\lambda\big)
$$ (H0:27)

```


$$
\begin{align}
& z_\mathrm{h}(t) = \left(u_0\,\cos(\omega_d\,t) +\frac{u_0\,\lambda  + v_0}{\omega_d}  \,\sin(\omega_d\,t)\right)\,e^{-\lambda\,t}
\\
& \dot{z}_\mathrm{h}(t) = \left(v_0  \,\cos(\omega_d\,t) - \frac{u_0\,\omega_0^2  + v_0\,\lambda}{\omega_d}\,\sin(\omega_d\,t)\right)\,e^{-\lambda\,t}
\\
&\ddot{z}_\mathrm{h}(t) = \left[a_0\,\,\cos(\omega_d\,t)-\frac{v_0\,\omega_0^2 + a_0\,\lambda }{\omega_d}  \,\sin(\omega_d\,t)\right]\,e^{-\lambda\,t}
\end{align}
$$ (H0:26)

$$
a_0 = -\big(u_0\,\omega_0^2  + 2\,v_0\,\lambda\big)
$$ (H0:27)

$$
\ddot{z}_\mathrm{h}(t) = \left(-v_0\,\omega_d\,\sin(\omega_d\,t) - ( u_0\,\omega_0^2  + v_0\,\lambda )\,\cos(\omega_d\,t) - v_0\,\lambda  \,\cos(\omega_d\,t) + \frac{u_0\,\omega_0^2  + v_0\,\lambda}{\omega_d}\,\lambda\,\sin(\omega_d\,t)\right)\,e^{-\omega_0\,\xi\,t}
$$

$$
\ddot{z}_\mathrm{h}(t) = \left( - ( u_0\,\omega_0^2  + 2\,v_0\,\lambda )\,\cos(\omega_d\,t)  + \frac{u_0\,\omega_0^2\,\lambda -v_0\,\omega_d^2 + v_0\,\lambda^2}{\omega_d}\,\sin(\omega_d\,t)\right)\,e^{-\omega_0\,\xi\,t}
$$

$$
\ddot{z}_\mathrm{h}(t) = \left( - ( u_0\,\omega_0^2  + 2\,v_0\,\lambda )\,\cos(\omega_d\,t)  + \frac{u_0\,\omega_0^2\,\lambda -v_0\,\omega_0^2 + 2\,v_0\,\lambda^2}{\omega_d}\,\sin(\omega_d\,t)\right)\,e^{-\omega_0\,\xi\,t}
$$

$$
\ddot{z}_\mathrm{h}(t) = \left( - ( u_0\,\omega_0^2  + 2\,v_0\,\lambda )\,\cos(\omega_d\,t)  + \frac{\lambda\,(u_0\,\omega_0^2+2\,v_0\,\lambda) -v_0\,\omega_0^2}{\omega_d}\,\sin(\omega_d\,t)\right)\,e^{-\omega_0\,\xi\,t}
$$

$$
\ddot{z}_\mathrm{h}(t) = \left( a_0\,\cos(\omega_d\,t)  - \frac{\lambda\,a_0 +v_0\,\omega_0^2}{\omega_d}\,\sin(\omega_d\,t)\right)\,e^{-\omega_0\,\xi\,t}
$$


```{div} full-width

Let us now implement eq. {eq}`H0:26` in python

```

`````python
import numpy as np

def damped_oscillator(ts, U, V, natural_period=1, damping_ratio=0.1):
    """
    \t\n
    Return the relative displacement `zh`, velocity `dzh` and acceleration `ddzh` 
    of a damped harmonic oscillator with initial condition `U` and `V` evaluated  
    at the times `ts`. 

    The natural period and damping ratio can be set as optional arguments.
    """
    
    w0  = 2*np.pi/natural_period
    wd  = w0 * np.sqrt(1-damping_ratio**2)
    lam = w0*damping_ratio
    
    A = - U*w0**2 - 2*V*lam
    
    relative_motions = np.empty((3,)+ts.shape)
    
    relative_motions[0] = ( U *np.cos(wd*ts) + (U*lam + V)/wd       * np.sin(wd*ts) ) *  np.exp(-lam*ts)
    relative_motions[1] = ( V *np.cos(wd*ts) - (U*w0**2 + V*lam)/wd * np.sin(wd*ts) ) *  np.exp(-lam*ts)
    relative_motions[2] = ( A *np.cos(wd*ts) - (V*w0**2 + A*lam)/wd * np.sin(wd*ts) ) *  np.exp(-lam*ts)
    
    return relative_motions
    
help(damped_oscillator)
`````

## General solution

```{div} full-width

In view of the fact that the differential equation of the driven harmonic oscillator is linear in both $z$ and $u$ and their derivatives, its general solution, for any ground acceleration, can be expressed in the following form of a time convolution 

$$
z(t) = \int_{-\infty}^\infty \ddot{u}_g(\tau)\,G(t-\tau)\,d\tau
$$ (H0:28)

between the ground acceleration and a Green function $G$ that we are going to determine.

```

### Green function

```{div} full-width

Substituting eq. {eq}`H0:28` into eq. {eq}`H0:6` or {eq}`H0:12` we obtain

$$
\int_{-\infty}^\infty \ddot{u}_g(\tau)\,\big(\ddot{G}(t-\tau)+2\,\lambda\,\dot{G}(t-\tau)+\omega_0^2\,G(t-\tau)\big)\,d\tau = -u_g(t)
$$ (H0:29)

From this point of view, we can understand that we are looking for a Green function $G$ such that

$$
\ddot{G}(t)+2\,\lambda\,\dot{G}(t)+\omega_0^2\,G(t) = - \delta(t)
$$ (H0:30)

In this perspective, we test the following trial solution

$$
G(t) = g(t)\,H(t)
$$ (H0:31)

with $g$ satisfying the following initial conditions

$$
g(0) = 0 \qquad \dot{g}(0) = -1
$$ (H0:32)

These initial conditions have been defined in such a way that

$$
\begin{align}
& \dot{G}(t) = \dot{g}(t)\,H(t) + g(0)\,\delta(t) =  \dot{g}(t)\,H(t)
\\
& \ddot{G}(t) = \ddot{g}(t)\,H(t) + \dot{g}(0)\,\delta(t) = \ddot{g}(t)\,H(t) -\delta(t)  
\end{align}
$$ (H0:33)

and, so, eq. {eq}`H0:30` simplifies into 

$$
\big(\ddot{g}(t)+2\,\lambda\,\dot{g}(t)+\omega_0^2\,g(t)\big)\,H(t) = 0
$$ (H0:34)

In this respect, $g$ is the homoegenous solution for the harmonic oscillator with initial conditions given by eq. {eq}`H0:32`. From eq. {eq}`H0:26` we thus obtain

$$
g(t) = -\frac{\sin(\omega_d\,t)\,e^{-\lambda\,t}}{\omega_d}
$$ (H0:35)
```

### Response function

```{div} full-width
Starting from the Green function, eqs. {eq}`H0:31` and {eq}`H0:35`, we can also obtain the response function $R$ relating the ground and relative accelerations

$$
\ddot{z}(t) = \int_{-\infty}^\infty \ddot{u}_g(\tau)\,\ddot{G}(t-\tau)\,d\tau = \int_{-\infty}^\infty \ddot{u}_g(\tau)\,R(t-\tau)\,d\tau
$$ (H0:36)

that is the second order derivative of $G$

$$
\begin{align}
R(t) &= \ddot{G}(t) = \ddot{g}(t)\,H(t) -\delta(t)
\\
&= \left[ \frac{\omega_d^2-\lambda^2}{\omega_d}\,\sin(\omega_d\,t) +2\,\lambda\,\cos(\omega_d\,t)\right]\,e^{-\lambda\,t}\,H(t) - \delta(t)
\end{align}
$$ (H0:37)

It is straightforward to show that the response function also relates the ground and relative displacements, as well as the ground and relative velocities


$$
z(t) = \int_{-\infty}^\infty u(\tau)\,R(t-\tau)\,d\tau
\qquad
\dot{z}(t) = \int_{-\infty}^\infty \dot{u}(\tau)\,R(t-\tau)\,d\tau
$$ (H0:38)

Although the Green and response functions allow to determine the relative ground motion, their application requires a time convolution that can be tedious to be computed analytically and hide the physical understanding of the problem. We will use them in the frequency domain after {ref}`chap-F` and to make practise on the {ref}`chap-D`.

```


## Sinusoidal ground acceleration


In order to understand how the harmonic oscillator behaves when driven by ground motion, let us consider the following sinusoidal ground acceleration with period $T$ and compact support on $[0,k\,T]$

$$
\ddot{u}_g = g(t)\,\big(H(t)-H(t-k\,T)\big)
$$ (H0:39)

with

$$
g(t) = A_g\,\sin(\alpha\,t)
$$

and $\alpha$ being the angular frequency associated to the ground period $T$

$$
\alpha = \frac{2\,\pi}{T}
$$ (H0:40)

As we are going to show, the relative motion can be expressed in the followinbg form

$$
z(t) = z_\mathrm{p}(t)\,\big(H(t)-H(t-k\,T)\big) + z_\mathrm{h,0}(t)\,H(t) - z_\mathrm{h,k}(t)\,H(t-k\,T)
$$ (H0:41)

where $z_\mathrm{h}$ is an homogeneous solution and $z_\mathrm{p}$ is a particular solution of 

$$
\ddot{z}_p(t)+2\,\lambda\,\dot{z}_p(t)+\omega_0\,z_\mathrm{p}(t) = -g(t)
$$ (H0:42)

Here the Heaviside function $H$ has been introduced because there is no motion before the beginning of the ground motion and it affects the first and second order derivatives as follow


$$
\begin{align}
\dot{z}(t) &= \dot{z}_\mathrm{p}(t)\,\big(H(t)-H(t-k\,T)\big) + \dot{z}_\mathrm{h,0}(t)\,H(t) - \dot{z}_\mathrm{h,k}(t)\,H(t-k\,T)
\\
&+ z_\mathrm{p}(t)\,\big(\delta(t)-\delta(t-k\,T)\big) + z_\mathrm{h,0}(t)\,\delta(t) - z_\mathrm{h,k}(t)\,\delta(t-k\,T)
\\
\ddot{z}(t) &= \ddot{z}_\mathrm{p}(t)\,\big(H(t)-H(t-k\,T)\big) + \ddot{z}_\mathrm{h,0}(t)\,H(t) - \ddot{z}_\mathrm{h,k}(t)\,H(t-k\,T)
\\
&+ 2\,\dot{z}_\mathrm{p}(t)\,\big(\delta(t)-\delta(t-k\,T)\big) + 2\,\dot{z}_\mathrm{h,0}(t)\,\delta(t) - 2\,\dot{z}_\mathrm{h,k}(t)\,\delta(t-k\,T)
\\
&+ z_\mathrm{p}(t)\,\big(\dot{\delta}(t)-\dot{\delta}(t-k\,T)\big) + 2\,z_\mathrm{h,0}(t)\,\dot{\delta}(t) - 2\,z_\mathrm{h,k}(t)\,\dot{\delta}(t-k\,T)
\end{align}
$$ (H0:43)


$$
\begin{align}
&\dot{z}(t) = (\dot{z}_\mathrm{p}(t)+\dot{z}_\mathrm{h}(t))\,H + (z_\mathrm{p}(t)+z_\mathrm{h}(t))\,\delta(t)
\\
&\ddot{z}(t) = (\ddot{z}_\mathrm{p}(t)+\ddot{z}_\mathrm{h}(t))\,H + 2\,(\dot{z}_\mathrm{p}(t)+\dot{z}_\mathrm{h}(t))\,\delta(t) + (z_p+z_h)\,\dot{\delta}(t)
\end{align}
$$ (H0:43)

Once substituded in eq. {eq}`H0:6`, we obtain

$$
\begin{align}
& 2\,\dot{z}_\mathrm{p}(t)\,\big(\delta(t)-\delta(t-k\,T)\big) + 2\,\dot{z}_\mathrm{h,0}(t)\,\delta(t) - 2\,\dot{z}_\mathrm{h,k}(t)\,\delta(t-k\,T)
\\
&+ z_\mathrm{p}(t)\,\big(\dot{\delta}(t)-\dot{\delta}(t-k\,T)\big) + 2\,z_\mathrm{h,0}(t)\,\dot{\delta}(t) - 2\,z_\mathrm{h,k}(t)\,\dot{\delta}(t-k\,T)
\\
&+ \lambda\,z_\mathrm{p}(t)\,\big(\delta(t)-\delta(t-k\,T)\big) + \lambda\,z_\mathrm{h,0}(t)\,\delta(t) - \lambda\,z_\mathrm{h,k}(t)\,\delta(t-k\,T)
\end{align}
$$ (H0:43)

$$
\begin{align}
& \delta(t)\,\Big[2\,\dot{z}_\mathrm{p}(t) + \lambda\,z_\mathrm{p}(t) + 2\,\dot{z}_\mathrm{h,0}(t)+\lambda\,z_\mathrm{h,0}(t)\Big] + \dot{\delta}(t)\,\Big[z_\mathrm{p}(t)+z_\mathrm{h,0}(t)\Big]
\\
& -\delta(t-k\,T)\,\Big[2\,\dot{z}_\mathrm{p}(t) + \lambda\,z_\mathrm{p}(t) + 2\,\dot{z}_\mathrm{h,k}(t)+\lambda\,z_\mathrm{h,k}(t)\Big] - \dot{\delta}(t-k\,T)\,\Big[z_\mathrm{p}(t)+z_\mathrm{h,k}(t)\Big] = 0
\end{align} 
$$ (H0:43)

$$
z_\mathrm{p}(0)+z_\mathrm{h,0}(0) = 0 \qquad
\dot{z}_\mathrm{p}(0)+\dot{z}_\mathrm{h,0}(0) = 0 \qquad 
z_\mathrm{p}(k\,T)+z_\mathrm{h,k}(k\,T) = 0 \qquad
\dot{z}_\mathrm{p}(k\,T)+\dot{z}_\mathrm{h,k}(k\,T) = 0
$$


$$
2\,(\dot{z}_\mathrm{p}(t)+\dot{z}_\mathrm{h}(t))\,\delta(t) + (z_\mathrm{p}(t)+z_\mathrm{h}(t))\,\dot{\delta}(t) + 2\,\lambda\,(z_\mathrm{p}(t)+z_\mathrm{h}(t))\,\delta(t) = 0
$$ (H0:44)

Within this framework, we can understand that eq. {eq}`H0:41` solves eq. {eq}`H0:6` before and after the initial time, within the time intervals $(-\infty,0)$ and $(0,\infty)$.  In order that it solve the differential equation also in the surrounding of the initial time $t=0$, we need to choose the homogeneous solution in such a way that $z$ and its first derivative are continuous at $t=0$, which means that we have to require that 

$$
z_\mathrm{p}(0)+z_\mathrm{h}(0) = 0 \qquad \dot{z}_\mathrm{p}(0) +\dot{z}_\mathrm{h}(0) = 0
$$ (H0:45)

We can fullfil this requirement by setting the initial displacement and velocity for the homoegenous solution have to be choosen as follows

$$
u_0 = -z_\mathrm{p}(0)  \qquad v_0 = -\dot{z}_\mathrm{p}(0) 
$$ (H0:46)

```{admonition} Transient and steady-state solutions
:class: note

Although we have not yet determined $z_\mathrm{p}$ and $z_\mathrm{h}$, we can already state that the latter describes a transient response of which the amplitude gradually decreases to zero with the exponential decay $\lambda$. The former, instead, describes the steady state solution that persists and will be the only one after the initial transient.

```



### Particular (or steady state) solution

````{div} full-width

Let us solve eq. {eq}`H0:42` starting from the following trial solution

$$
z_p(t) = - A_g\,\frac{A}{\alpha^2}\,\sin(\alpha\,t+\beta)
$$ (H0:47)

The relative velocity and acceleration read

$$
\dot{z}_p(t) = - A_g\,\frac{A}{\alpha}\,\cos(\alpha\,t+\beta)
\qquad \ddot{z}(t) = A_g\,A\,\sin(\alpha\,t+\beta) 
$$ (H0:48)

and, so, $A$ and $\beta$ can be considered as the amplitude factor and phase shift of the relative acceleration with respect to the ground acceleration. Also, we note that we can write the relative acceleration as follows

$$
\ddot{z}_p(t) = A_g\,A\,\sin(\alpha\,(t+\delta t)) 
$$ (H0:49)

where $\delta t$ is the shift in time $\delta t = \beta/\alpha$.

Substituting eqs. into eq., making use of the following trignognometric identities 

$$
\begin{align}
\sin(\alpha\,t+\beta) = \sin(\alpha\,t)\,\cos\beta + \cos(\alpha\,t)\,\sin\beta
\\
\cos(\alpha\,t+\beta) = \sin(\alpha\,t)\,\sin\beta - \cos(\alpha\,t)\,\cos\beta
\end{align}
$$ (H0:50)

and collecting the sine, $\sin(\alpha\,t)$, and cosine, $\cos(\alpha\,t)$, terms, we obtain

$$
-\frac{A\,\cos(\alpha\,t)}{\alpha^2}\,\left(\big(\omega_0^2-\alpha^2\big)\,\sin\beta + 2\,\alpha\,\lambda\,\cos\beta\right)
-\frac{A\,\sin(\alpha\,t)}{\alpha^2}\,\left(\big(\omega_0^2-\alpha^2\big)\,\cos\beta - 2\,\alpha\,\lambda\,\sin\beta\right) = - \sin(\alpha\,t)
$$ (H0:51)

Just because the right-hand side is characterized by the only sine term, the first term in left-hand side have to be zero and we can do it choosing

$$
\cos\beta = \frac{\omega_0^2-\alpha^2}{D} \qquad \sin\beta = \frac{-2\,\alpha\,\lambda}{D}
$$ (H0:52)

with

$$
D = \sqrt{\big(\omega_0^2-\alpha^2)^2+(2\,\alpha\,\lambda)^2}
$$ (H0:53)

With this choice, eq. {eq}`H0:51` becomes

$$
-\frac{A\,D\,\sin(\alpha\,t)}{\alpha^2} = - \sin(\alpha\,t)
$$ (H0:54)

that is satisfied by

$$
A=\frac{\alpha^2}{D}
$$ (H0:55)

```{admonition} Resonance frequency
:class: note

Let us investigate the frequency at which the amplitude factor $A$ reaches its maximum value. By investigating the stationary points

$$
\frac{\partial A}{\partial \alpha} = \frac{2\,\alpha\,\omega_0^2}{D^3}\,\big[\omega_0^2-\alpha^2\,\big(1-2\,\xi^2\big)\big]
$$ (H0:56)

we found that $\alpha=0$ is a minimum, while 

$$
\alpha = \omega_0\,\sqrt{1-2\,\xi^2}
$$ (H0:57)

is a maximum. The latter, however, there exists only when $\xi<1/\sqrt{2}\approx 0.707$.
```

````

```{div} full-width

Let us now implement eqs. {eq}`H0:47` and {eq}`H0:48` for compunting the steady state relative motion due to the sinusoidal ground acceleration

```

`````python
def setup_driven_oscillator(alpha, natural_period=1, damping_ratio=0.1):

    w0 = 2*np.pi/natural_period
    lam = w0*damping_ratio
    
    D = np.sqrt((w0**2-alpha**2)**2+(2*alpha*lam)**2)
    C = (w0**2-alpha**2)/D
    beta = - np.arccos(C)
    amplitude_factor = alpha**2/D
    
    #fmax = 1/natural_period/np.sqrt(1-2*damping_ratio**2)

    return amplitude_factor, beta

def driven_oscillator(ts, ground_period=1, natural_period=1, damping_ratio=0.1):
    
    alpha = 2*np.pi/ground_period
    amplitude_factor, beta = setup_driven_oscillator(alpha,natural_period,damping_ratio)
    
    ground_acceleration = np.sin(alpha*ts)  
    relative_motions = np.empty((3,)+ts.shape)
    relative_motions[0] = - amplitude_factor / alpha**2 * np.sin(alpha*ts+beta)
    relative_motions[1] = - amplitude_factor / alpha    * np.cos(alpha*ts+beta)
    relative_motions[2] =   amplitude_factor            * np.sin(alpha*ts+beta)
    return ground_acceleration, relative_motions
`````


### Transient response

```{div} full-width

Having obtained the particular (steady-state) relative motion, we can now determine the associated transient response with the initial conditions given by eq. {eq}`H0:46` making use of eqs. {eq}`H0:47` and {eq}`H0:48`. In particular, the transient response $z_\mathrm{h}$ can be obtained from eq. {eq}`H0:26` using 

$$
u_0 = -z_\mathrm{p}(0) = \frac{A}{\alpha^2}\,\sin\beta \qquad v_0 = -\dot{z}_\mathrm{p}(0) = \frac{A}{\alpha}\,\cos\beta
$$ (T:58)

as initial conditions.

We can now implement in python both the steady-state and transient solutions
```

`````python
def driven_oscillator_heaviside(ts, ground_period=1, natural_period=1, damping_ratio=0.1):
    
    # setup 
    alpha = 2*np.pi/ground_period
    amplitude_factor, beta = setup_driven_oscillator(alpha,natural_period,damping_ratio)
    U = amplitude_factor/alpha**2 * np.sin(beta)
    V = amplitude_factor/alpha    * np.cos(beta)
    
    # the homogeneous (or transient) solution
    transient_motions = damped_oscillator(ts, U, V, natural_period, damping_ratio)
    
    # the particular solution
    ground_acceleration,relative_motions = driven_oscillator(ts, ground_period, natural_period, damping_ratio)

    # implementing eq. ()
    relative_motions += transient_motions

    # implementing the Heaviside function
    ground_acceleration[ts<0] = 0
    transient_motions[:,ts<0] = 0
    relative_motions[:,ts<0]  = 0
    
    return ground_acceleration, relative_motions, transient_motions
`````

```{div} full-width
and plot the comparison between the ground and relative motion, as well as visualize the contribution from the transient motion
```

## Test function

The sinusoidal ground acceleration, although usefull for a first understanding of the bahaviour of the driven harmonic oscillator, does not describe a typical seismic wave.

Considring the ground velocity and displacement, indeed, we have

$$
\dot{u}_g(t) = \int_{-\infty}^t \ddot{u}_g(\tau)\,d\tau =  H(t)\,\int_{0}^t \sin(\alpha\,\tau)\,d\tau = A_g\,\left(\frac{1-\cos(\alpha\,t)}{\alpha}\right)\,H(t)
$$

$$
u_g(t) = \int_{-\infty}^t \dot{u}_g(\tau)\,d\tau =  A_g\,H(t)\,\int_{0}^t \left(\frac{\cos(\alpha\,\tau)-1}{\alpha}\right)\,d\tau = \frac{1}{\alpha}\,\left(t-\frac{\sin(\alpha\,t)}{\alpha}\right)\,H(t)
$$

and we can see that the ground displacement is characterized by a linear trend beginning at $t=0$.

A smooth function that can be used as seismic wavelet packet is

$$
\begin{align}
& \ddot{u}_g(t) = A_g\,\big[\cos(\alpha\,t) - \cos(2\,\alpha\,t)\big]\,B_{A}(t)
\\
& \dot{u}_g(t) = \frac{A_g}{\alpha}\,\big[\sin(\alpha\,t)-\tfrac{1}{2}\,\sin(2\,\alpha\,t))\big]\,B_{A}(t)
\\
& u_g(t) = \frac{A_g}{\alpha^2}\,\big[\tfrac{1}{4}\,\cos(2\,\alpha\,t)-\cos(\alpha\,t) + \tfrac{3}{4}\big]\,B_{A}(t)
\end{align}
$$ 

where the angular frequency $\alpha$ associated to the period $A$

$$
\begin{align}
\alpha = \frac{2\,\pi}{A}
\end{align}
$$ 


$$
\ddot{z}_p(t) = A_g\,\big[A(\alpha)\,\cos\big(\alpha\,t+\beta(\alpha)\big) - A(2\,\alpha)\,\cos\big(2\,\alpha\,t+\beta(2\,\alpha)\big)
$$

$$
\dot{z}_p(t) = A_g\,\big[\frac{A(\alpha)}{\alpha}\,\sin\big(\alpha\,t+\beta(\alpha)\big) - \frac{A(2\,\alpha)}{2\,\alpha}\,\sin\big(2\,\alpha\,t+\beta(2\,\alpha)\big)
$$

$$
z_p(t) = A_g\,\big[-\frac{A(\alpha)}{\alpha^2}\,\cos\big(\alpha\,t+\beta(\alpha)\big) + \frac{A(2\,\alpha)}{(2\,\alpha)^4}\,\cos\big(2\,\alpha\,t+\beta(2\,\alpha)\big)
$$

$$
\ddot{z}_p(k\,T_g) = A_g\,\big[A(\alpha)\,\cos\big(\beta(\alpha)\big) - A(2\,\alpha)\,\cos\big(\beta(2\,\alpha)\big)
$$

$$
\dot{z}_p(k\,T_g) = A_g\,\big[\frac{A(\alpha)}{\alpha}\,\sin\big(\beta(\alpha)\big) - \frac{A(2\,\alpha)}{2\,\alpha}\,\sin\big(\beta(2\,\alpha)\big)
$$

$$
z_p(k\,T_g) = A_g\,\big[-\frac{A(\alpha)}{\alpha^2}\,\cos\big(\beta(\alpha)\big) + \frac{A(2\,\alpha)}{(2\,\alpha)^4}\,\cos\big(\beta(2\,\alpha)\big)
$$

`````python
def test_function(ts,A):
    alpha = 2*np.pi/A
    ground_motions  = np.empty((3,)+ts.shape)
    x = alpha*ts
    ground_motions[0] = (   np.cos(2*x)/4 - np.cos(x) + 3/4 ) / alpha**2
    ground_motions[1] = ( - np.sin(2*x)/2 + np.sin(x)       ) / alpha
    ground_motions[2] = ( - np.cos(2*x)   + np.cos(x)       )
    ground_motions[:,ts<0] = 0
    ground_motions[:,ts>A] = 0
    return ground_motions
`````

\dot{z}_p(k\,T_g) = A_g\,\big[\frac{A(\alpha)}{\alpha}\,\sin\big(\beta(\alpha)\big) - \frac{A(2\,\alpha)}{2\,\alpha}\,\sin\big(\beta(2\,\alpha)\big)
$$

$$
z_p(k\,T_g) = A_g\,\big[-\frac{A(\alpha)}{\alpha^2}\,\cos\big(\beta(\alpha)\big) + \frac{A(2\,\alpha)}{(2\,\alpha)^4}\,\cos\big(\beta(2\,\alpha)\big)
$$

`````python
def driven_test_function(ts, ground_period=1, natural_period=1, damping_ratio=0.1):
    
    # setup 
    alpha = 2*np.pi/ground_period
    alpha2 = 2*alpha
    
    A, beta = setup_driven_oscillator(alpha,natural_period,damping_ratio)
    U = A/alpha**2 * np.cos(beta)
    V = - A/alpha    * np.sin(beta)
    
    A2, beta2 = setup_driven_oscillator(alpha2,natural_period,damping_ratio)
    U += - A2/alpha2**2 * np.cos(beta2)
    V += A2/alpha2    * np.sin(beta2)
    
    x  = alpha  * ts + beta
    x2 = alpha2 * ts + beta2
    
    relative_motions = np.empty((3,)+ts.shape)
    relative_motions[0] = ( - A*np.cos(x) + A2*np.cos(x2)/4 ) / alpha**2
    relative_motions[1] = (   A*np.sin(x) - A2*np.sin(x2)/2 ) / alpha
    relative_motions[2] =     A*np.cos(x) - A2*np.cos(x2)
    
    relative_motions[:,ts<0]  = 0
    relative_motions[:,ts>ground_period]  = 0

    # the homogeneous (or transient) solution
    transient_motions  = damped_oscillator(ts, U, V, natural_period, damping_ratio)
    transient_motions2 = damped_oscillator(ts-ground_period, U, V, natural_period, damping_ratio)
    transient_motions[:,ts<0] = 0
    transient_motions2[:,ts<ground_period] = 0
    transient_motions -= transient_motions2
    
    relative_motions += transient_motions

    return relative_motions, transient_motions
`````

`````python

%matplotlib widget
relative_motions, transient_motions = driven_test_function(ts,A)

ylabels = ["Displacement [m]","Velocity [m/s]","Acceleration [m/s$^2$]"]

fig,axes = plt.subplots(3,2,tight_layout=True,figsize=(10,6),sharex=True)
for ax in axes[-1]:
    ax.set_xlabel("Time [s]")
for axe,ground_motion,relative_motion,transient_motion,ylabel in zip(axes,ground_motions,relative_motions,transient_motions,ylabels):
    ax = axe[0]
    ax.plot(ts,ground_motion,label="ground")
    ax.plot(ts,relative_motion,label="relative")
    ax.set_ylabel(ylabel)
    ax.axhline(0,**options)
    ax.legend()    
    ax = axe[1]
    ax.plot(ts,relative_motion-transient_motion,label="steady-state")
    ax.plot(ts,transient_motion,label="transient")
    ax.axhline(0,**options)
    ax.legend()
`````
