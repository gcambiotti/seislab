```python
%matplotlib widget
import lab
import numpy as np
import matplotlib.pyplot as plt

from obspy.core.event import read_events
from obspy.core.inventory import read_inventory

directory = "Amatrice-Norcia-Visso/"

file = directory+"Seismic_Inventory.xml"
inventory = read_inventory(file)

file = directory + "7073641_Accumuli.xml"
catalog = read_events(file)

event = catalog[0]
origin = event.preferred_origin()

directory += "7073641_Accumuli/"
```

```python
import glob

files = glob.glob(directory+"*")
print(files)
```

```python
from geopy import distance as geopy_distance

mydict = {}
ids = []
for file in files:
    id = file.split("/")[-1][:-6]
    codes =id.split(".")
    channel = codes[-1]
    instrument = channel[1]
    if instrument not in ["L","N"]: continue
    inv = inventory.select(*codes)
    station = inv[0][0]
    distance = geopy_distance.distance((station.latitude,station.longitude),(origin.latitude,origin.longitude)).km
    mydict[id] = distance
```

```python
keys = np.array(list(mydict.keys()))
distances = np.array(list(mydict.values()))
s = np.argsort(distances)
distances = distances[s]
keys = keys[s]
mydict = {}
for key,distance in zip(keys,distances):
    mydict[key] = distance
```

```python
def eva_FT_elastic_acc(Ts,damping_ratio=0.1,F=3):
    eta = np.sqrt(10/(5+damping_ratio*100))
    TC = 0.2
    TB = TC/3
    TD = 1.6
    spectrum = np.ones(Ts.shape)*eta*F
    whe = Ts <= TB
    T = Ts[whe]
    spectrum[whe] *= T/TB+(1-T/TB)/(eta*F)
    whe = np.logical_and(Ts>TC,Ts<=TD)
    T = Ts[whe]
    spectrum[whe] *= TC/T
    whe = Ts>TD
    T = Ts[whe]
    spectrum[whe] *= TC*TD/T**2
    return spectrum
       
```
